```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import mlrose
import time
import NNOptimizer
from sklearn.preprocessing import StandardScaler, LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
data = pd.read_csv("BreastCancerData.csv")
```

```python
# Normalize x
x = data.drop(['id', 'diagnosis','Unnamed: 32'], axis='columns')
scaler = StandardScaler()
scaler.fit(x)
x1 = scaler.transform(x.values)
x1 = pd.DataFrame(data=x1)
x1.columns = x.columns
# Normalize y
y = data['diagnosis']
binarizer = LabelBinarizer()
y1 = binarizer.fit_transform(y).flatten()
y1 = pd.Series(data=y1)
# Split
X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size = 0.2)
result = {}
```

```python
max_attempts_range = [10, 20, 50]
restarts_range = [0, 10, 20]
result['RC'] = NNOptimizer.NN_RC_report(max_attempts_range, restarts_range, X_train, X_test, y_train, y_test)
```

```python
pd.DataFrame(result['RC']).to_csv('RC.csv')
```

```python
max_attempts_range = [50, 100]
init_temp_range = [1, 5]
decay = 0.99
temp_min_range = [0.01, 0.1]
result['SA'] = NNOptimizer.NN_SA_report(max_attempts_range, init_temp_range, decay, temp_min_range,
                                           X_train, X_test, y_train, y_test)
```

```python
pd.DataFrame(result['SA'])
```

```python
pd.DataFrame(result['SA']).to_csv('SA.csv')
```

```python
max_attempts_range = [10, 50]
pop_size_range = [10, 50]
mutation_prob_range = [0.1, 0.3]
result['GA'] = NNOptimizer.NN_GA_report(max_attempts_range, pop_size_range, mutation_prob_range,
                                           X_train, X_test, y_train, y_test)
```

```python
pd.DataFrame(result['GA']).to_csv('GA.csv')
```
